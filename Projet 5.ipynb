### un test d'adéquation : parmi les 4 variables, ou parmi d'autres variables que vous trouverez pertinentes, trouvez une variable dont la loi est normale.


```r
#Test de l'adéquation des variables à une loi normale
variable_test = df_complet$import_volaille_kg_pers_an

ks.test(variable_test,"pnorm",mean=mean(variable_test),sd=sd(variable_test))
```

p-value = 1.534e-11, il est donc possible de rejeter l'hypothése H0, de normalité. La P valeur est bien inférieure aux 5% de l'intervale de confiance

```r
variable_test = df_complet$pib_par_habitant_dollar

ks.test(variable_test,"pnorm",mean=mean(variable_test),sd=sd(variable_test))
```

p-value = 0.00011, il est donc possible de rejeter l'hypothése de normalité

```r
variable_test = df_complet$disp_Kcal_pers_j

ks.test(variable_test,"pnorm",mean=mean(variable_test),sd=sd(variable_test))
```

p-value = 0.7379, on ne peut donc pas rejeter l’hypothèse de normalité au niveau de test 5%.
La variable disp_Kcal_pers_j suit sans aucun doute une loi normale, 

### un test de comparaison de deux populations (dans le cas gaussien) : choisissez 2 clusters parmi ceux que vous aurez déterminé. Sur ces 2 clusters, testez la variable gaussienne grâce à un test de comparaison.

```r
#Je teste la population sur le cluster 2 et 3

clust2 <- df_complet[df_complet$cluster==2,]$diff_pop_en_p
clust3 <- df_complet[df_complet$cluster==3,]$diff_pop_en_p

#Test des variances
var.test(clust2,clust3)
```

Nous ne pouvons pas rejeter H0

```r
#On teste ensuite l’égalité des moyennes
t.test(clust2,clust3,var.equal=TRUE)
```

On ne rejette pas l'hypothèse à un niveau de test de 5%. Ces deux clusters suivent bien les mêmes lois et ce sont bien deux échantillons Gaussien
